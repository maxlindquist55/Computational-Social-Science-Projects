```python
#import packages for deep learning
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf

from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
#from keras.utils import np_utils
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import RMSprop
from transformers import pipeline
```

```python
# Here, 'text-generation' tells the pipeline to create a text generation model, and model='gpt2'
# specifies that you want to use the GPT-2 model for this task.
generator = pipeline('text-generation', model='gpt2')
```

```python
#set_seed(100)
input_text = "This project uses natural language processing techniques to explore a dataset containing tweets from members of the 116th United States Congress that met from January 3, 2019 to January 2, 2021. This is important because"
# specifying the maximum length of the output, and the number of different sequences you want
output = generator(input_text, max_length=100, num_return_sequences=5,temperature = 0.5)
for result in output:
    print(result['generated_text'])
    print("----")
```
