```python
#
# Recode variables
#-----------

# Diabetes
# ----------
# keep only useful info about our target feature, i.e., where diabetes_number not = 'Suppressed'
# note that the inside reference to the diabetes df identifies the column, and the outer calls specific rows according to a condition 
diabetes = diabetes[diabetes['Diabetes_Number']!="Suppressed"] 

# use the astype method on Diabetes_Number to convert it to an integer...if you are not sure, what does the astype() documentation tell you are possible arguments? 
diabetes['Diabetes_Number'] = diabetes['Diabetes_Number'].astype('int') 

# Obesity
# ----------
# keep only useful info about our target feature, i.e., where Obesity_Number not = 'No Data'
# note that the inside reference to the diabetes df identifies the column, and the outer calls specific rows according to a condition 
diabetes = diabetes[diabetes['Obesity_Number']!="No Data"] 
diabetes['Obesity_Number'] = diabetes['Obesity_Number'].astype('int') 


# Physical Inactivity
# ----------
# keep only useful info about our target feature, i.e., where Physical_Inactivity_Number not = 'No Data'
diabetes = diabetes[diabetes['Physical_Inactivity_Number']!="No Data"] 
diabetes['Physical_Inactivity_Number'] = diabetes['Physical_Inactivity_Number'].astype('int') 


# Some final changes 
# ----------
# 65+ sex ratio had one "-" in it so let's drop that row first
diabetes = diabetes[diabetes['sex and age_total population_65 years and over_sex ratio (males per 100 females)']!= "-"]

# change to numeric from string, since it originally included the "-", which made it a string
# you'll have to decide whether to make it integer or float 
diabetes['sex and age_total population_65 years and over_sex ratio (males per 100 females)'] = diabetes['sex and age_total population_65 years and over_sex ratio (males per 100 females)'].astype('float')
```

#### d. Check for duplicate columns

There are a lot of columns in this data frame. Let's see if there are any are duplicates. Note that Pandas will not allow them to have the same exact column name, so they will likely be distinct on column name but will be copies otherwise.

```python
#
# Check for duplicate columns
#-----------
# I used Google to figure this out, and adapted this example for our purposes:  
# source: https://thispointer.com/how-to-find-drop-duplicate-columns-in-a-dataframe-python-pandas/ 
def getDuplicateColumns(df):
    '''
    Get a list of duplicate columns.
    It will iterate over all the columns in dataframe and find the columns whose contents are duplicate.
    :param df: Dataframe object
    :return: List of columns whose contents are duplicates.
    '''
    duplicateColumnNames = set()
    # Iterate over all the columns in dataframe
    for x in range(df.shape[1]):
        # Select column at xth index.
        col = df.iloc[:, x]
        # Iterate over all the columns in DataFrame from (x+1)th index till end
        for y in range(x + 1, df.shape[1]):
            # Select column at yth index.
            otherCol = df.iloc[:, y]
            # Check if two columns at x 7 y index are equal
            if col.equals(otherCol):
                duplicateColumnNames.add(df.columns.values[y])
    return list(duplicateColumnNames)

duplicateColumnNames = list(getDuplicateColumns(diabetes))
print('Duplicate Columns are as follows: ')
duplicateColumnNames
```

```python
#
# drop columns from duplicates list
#-----------
# now drop list of duplicate features from our df using the .drop() method
diabetes = diabetes.drop(columns=duplicateColumnNames) 
```

### 3.2 Partition Data and Standardization

Now, we will partition our data to prepare it for the training process. Ultimately we want to use a 60% train—20% validation—20% test in this case. More data in the training set lowers bias, but then increases variance in the validation/test sets. Balancing between bias and variance with choice of these set sizes is important as we want to ensure that there is enough data to train on to get good predictions, but also want to make sure our hold-out sets are representative enough.

Work through partitioning the data into the test/train/validation sets in the chunks below. Be sure to that if you are using Ridge or LASSO, you Standardize the data. Where you do this in the workflow matters so be clear about where you are doing this and why. 

```python
#
# Partition data
#-----------

# import library
from sklearn.model_selection import train_test_split

# create y dataframe 
y = diabetes_clean.Diabetes_Number

# create X dataframe (include everything except "Diabetes_Number", our target, 
# and "race alone or in combination with one or more other races_total population")
X =  diabetes_clean.drop(["Diabetes_Number", 
               "race alone or in combination with one or more other races_total population",
                         ], # list of variables to drop
               axis= 1)   # which axis do we want to specify here
```

```python
#
# Training/test split
#-----------

# set the random seed
np.random.seed(10)


# split the data so that it returns 4 values: X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X,                 # specify training dataset
                                                    y,                 # specify test dataset
                                                    train_size=.8,      # specify proportional split for training
                                                    test_size=.2)       # specify proportional split for test
```

```python
#
# Validation split
#-----------

# train_test_split returns 4 values: X_train, X_test, y_train, y_test, so how do we create a 60-20-20 train-validate-test split? 
X_train, X_validate, y_train, y_validate = train_test_split(X_train, 
                                                            y_train,
                                                            train_size=0.75, 
                                                            test_size=0.25)
```

```python
#
# Standardization
#-----------
# Given that we want to only standardize non-dichotomous variables, we need to find a 
# solution that will loop over only the columns we want to standardize. The code below
# identifies all non-dichotomous variables in our dataset and only standardizes those.

# load library and create instance of Standard Scaler 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()


# identify non-dichotomous columns we want to transform
columns = list(X_test.select_dtypes(include=['number']).loc[:, X_test.nunique() > 2])

# use loop to transform training data for only columns we want to transform
for column in columns:
    X_train[column] = scaler.fit_transform(X_train[column].values.reshape(-1, 1)).flatten()

# use loop to transform validation data for only columns we want to transform
for column in columns:
    X_validate[column] = scaler.fit_transform(X_validate[column].values.reshape(-1, 1)).flatten()

# use loop to transform test data for only columns we want to transform
for column in columns:
    X_test[column] = scaler.fit_transform(X_test[column].values.reshape(-1, 1)).flatten()
```

**Explanation**

I chose a 80/20 split for my training and testing datasets, and then a subsequent 75/25 split for my training and validation training set. This common splits strike a balance between having sufficient training data and a reasonably sized validation set. A very small validation set may lead to unreliable model selection, while a small training set might result in model underfitting.

I standardized my models after creating the training, validation, and test datasets in order to prevent data leakage.

### 4.1 Describe models

**MODEL DESCRIPTION(S):** 


***Model One -- Ordinary Least Squares***

**Basic Logic:** Ordinary Least Squares is a simple regression technique that models the linear relationship between outcomes and features by finding the best-fitting straight line that explains the variation between those two variables.

**Assumptions:** Ordinary Least Squares models assume linear relationship between features and outcomes, homoscedasticity, and no multicollinearity between features.

**Pros:** Ordinary Least Squares models are simple and easy to understand. 

**Cons:** Ordinary Least Squares models are very to underfitting due to their simplicity. They are also sensative to outliers and cannot model nonlinear or multicollinear relationships between features and outcomes.

**Plausible utility:** This is likely going to be the least plausible model for policymakers to use in this situation because it risks oversimplifying the relationship between diabetes, region, race, age, obesity, and inactivity. Similarly, the features in this dataset are likely colinear, which presents a problem for Ordinary Least Squares.


***Model Two -- Ridge Regression***

**Basic Logic:** Ridge Regression addresses some of the problems of Ordinary Least Squares by imposing a penalty on the size of the coefficients, which is called the L2 regularization term. This penalty term prevents the coefficients from becoming too large, which mitigates issues related to multicollinearity and overfitting. 

**Assumptions:** Ridge Regression still assumes a linear relationship between features and outcomes, homoscedasticity, and no endogeneity. 

**Pros:** Unlike Ordinary Least Squares, Ridge Regression can handle multicollinearity.

**Cons:** Unlike LASSO, Ridge Regression does not perform feature selection, which means irrelevant features will be included in the model.

**Plausible utility:** This model seems to occupy the mid-ground between Ordinary Least Squares and LASSO. While Ridge Regression may include insignificant features, we can still gauge the significance of diabetic risk factors without having to worry about the L1 parameter zeroing out coefficients.



***Model Three -- LASSO***

**Basic Logic:**  LASSO is a regression technique that addresses the limitations of Ordinary Least Squares by adding an L1 regularization term to the linear regression model. This L1 regularization imposes a penalty on the size of the coefficients, which, unlike the L2 regularization term in Ridge Regression, many zero out coefficients.

**Assumptions:** LASSO still assumes a linear relationship between features and outcomes, homoscedasticity, and no endogeneity. 

**Pros:** LASSO is very useful for feature selection, which can help policymakers target populations, regions, or lifestyle/environmental factors that are significant contributors to diabetes.

**Cons:** The major drawback with LASSO is that the model's L1 parameter risks setting coefficients that cause multicollinearity to zero. This is an issue if we one of the zeroed features is one of our main treatments of interest.

**Plausible utility:** This model grants policymakers the ability to target features that contribute to diabetes, while minimizing varience and bias.

We can determine if a model is over- or underfitting a dataset by testing it on both the validation and test sets. 1. What is bias-variance tradeoff? Why is it relevant to machine learning problems like this one?

**Response**

Bias-variance tradeoff refers to the problem in machine learning in which the practicioner needs to select a model that accurately captures the often-times complex underlying relationships between predictors and outcomes, while avoiding capturing irrelevant noise. Models that are biased are oversimplified models that fit neither the training nor testing data. Biased models are at risk for underfitting, which means that the model does not capture the relationship between features and outcomes. Models with high levels of variance cause overfitting, which occurs when the model closely fits the training data but has high error on the testing data. This higher error resutling from testing dataset is a result of the model capturing noise while being fitted to the training dataset.

Bias-variance tradeoff is important to the methods used in this project. The OLS model underfitted the dataset relative to the Ridge Regression and LASSO models.

As stated above, overfitting occurs when the model closely fits the training data but has high error on the testing data. This higher error resutling from testing dataset is a result of the model capturing noise while being fitted to the training dataset.
